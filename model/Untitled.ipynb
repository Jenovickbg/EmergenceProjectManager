```python
import csv
import random
from datetime import datetime, timedelta

noms_porteurs = ["Kabasele Group", "Jeunes Entrepreneurs Kin", "Femmes d'Impact", "StartUp RDC", "TechnoPlus", "AGRI-DEV", "Médicorp", "Green Congo"]
destinataires = ["Emergence SARL", "Ministère de l'Économie", "Investisseurs Privés", "Incubateur Émergence", "ONG Partenaire"]
localisations = ["Kinshasa", "Lubumbashi", "Goma", "Matadi", "Kisangani", "Bukavu", "Mbuji-Mayi"]
beneficiaires = ["PME locales", "jeunes diplômés", "femmes rurales", "populations urbaines", "étudiants", "centres de santé"]
domaines = ["NTIC", "SANTE", "AGRICULTURE", "CONSTRUCTION", "FORMATION", "BIOMÉTRIE", "FINTECH", "ÉDUCATION", "ÉNERGIE", "TOURISME"]
conclusions = [
    "Nous espérons un accompagnement stratégique.",
    "Ce projet changera positivement la vie des bénéficiaires.",
    "Il nécessite un soutien technique et financier.",
    "Sa mise en œuvre rapide est cruciale pour la zone ciblée.",
    "Nous sollicitons une analyse approfondie par Emergence."
]

def random_date():
    start_date = datetime(2024, 1, 1)
    end_date = datetime(2025, 6, 15)
    delta = end_date - start_date
    random_days = random.randint(0, delta.days)
    return (start_date + timedelta(days=random_days)).strftime("%d/%m/%Y")

# Générer et écrire le fichier
with open("projets_100_complets.csv", mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file, quoting=csv.QUOTE_ALL)
    writer.writerow([
        "titre", "date_soumission", "porteur_projet", "destinataire",
        "resume_projet", "objectifs_projet", "description_projet",
        "localisation", "beneficiaires", "budget_previsionnel", "conclusion"
    ])
    
    for i in range(100):
        domaine = random.choice(domaines)
        titre = f"Projet de {domaine.lower().capitalize()} #{i+1}"
        writer.writerow([
            titre,
            random_date(),
            random.choice(noms_porteurs),
            random.choice(destinataires),
            f"Ce projet vise à renforcer le secteur de {domaine.lower()} dans la région de {random.choice(localisations)}.",
            f"Améliorer les conditions dans le domaine de {domaine.lower()} pour les {random.choice(beneficiaires)}.",
            f"Activités : formation, dotation en équipements, sensibilisation communautaire dans {domaine.lower()}.",
            random.choice(localisations),
            random.choice(beneficiaires),
            f"{random.randint(5000, 200000)} USD",
            random.choice(conclusions)
        ])
```
